```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## During August 2018, severe floods affected the south Indian state Kerala,
## due to unusually high rainfall during the monsoon season.
 


## It was the worst flood in Kerala in nearly a century. Over 483 people died, and 140 are missing.

## We will review some important data on that incident

```python
df = pd.read_csv('./district_wise_details.csv')
df.plot()
```

```python
sns.set()
```

## This graph indicates the percentage of precipitation relative to normal time

```python
x = df['actual_rainfall_in_mm'].values
y = df['normal_rainfall_in_mm'].values
plt.plot(x , color = 'red')
plt.plot(y , color = 'blue')
```

## These graphics indicate the extent of the damage done

```python
x = df['actual_rainfall_in_mm'].values
y = df['full_damaged_houses'].values
plt.plot(x , color = 'red')
plt.plot(y , color = 'blue')
```
